```python
df = pd.read_csv("covid19/covid_19_data.csv")
df = df.groupby(by=["Country/Region", "ObservationDate"]).sum()
df.reset_index(inplace=True)#, level="Country/Region")
try:
    df.drop(["SNo"], axis=1, inplace=True)
except: 
    KeyError
    
df["Deaths/Confirmed"] = df["Deaths"] / df["Confirmed"]
df["Recovered/Confirmed"] = df["Recovered"] / df["Confirmed"]
df["Deaths/Recovered"] = df["Deaths"] / df["Recovered"]

df["Closed Cases"] = df["Deaths"] + df["Recovered"] 
df["Recovery Rate"] = df["Recovered"]  / df["Closed Cases"]
df["Death Rate"] = df["Deaths"]  / df["Closed Cases"]

df.sort_values(by=['ObservationDate'], inplace=True)
#df
df.rename(columns={"Country/Region":"Country"}, inplace=True)
```

```python
df.groupby("Country").max()
df.sort_values(by=['Deaths'],ascending=False, inplace=True)
df_bool_duplicated = df.duplicated(subset="Country", keep="first")
#df.sort_values(by=['ObservationDate'],ascending=True)
max_deaths = df[~(df_bool_duplicated)]
max_deaths.reset_index(drop=True, inplace=True)
#max_deaths.head()
```

```python
max_deaths_10 = max_deaths.loc[0:9]
max_deaths_10_list = max_deaths_10["Country"].to_list()

max_deaths_5 = max_deaths.loc[0:4]
max_deaths_5_list = max_deaths_5["Country"].to_list()
```

```python
df_top10 = df.filter(items=max_deaths_10_list, axis="index")
df_top10= df[df["Country"].isin(max_deaths_10_list)]
df_top10 = df_top10.sort_values(by=['ObservationDate'],ascending=True)

df_top5 = df.filter(items=max_deaths_5_list, axis="index")
df_top5= df[df["Country"].isin(max_deaths_5_list)]
df_top5 = df_top5.sort_values(by=['ObservationDate'],ascending=True)

#df_top10
```
